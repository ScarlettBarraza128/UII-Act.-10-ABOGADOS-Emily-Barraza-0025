## Abogado

```python
# ==============================================
# IA DE ABOGADOS VIRTUALES
# Uso de:
# Lista de diccionarios
# if, elif, else
# while True
# ciclo for
# ==============================================

# Lista vacía al inicio
losabogados = []

print("====================================")
print("     IA DE ABOGADOS VIRTUALES")
print("====================================")

# Ciclo principal
while True:

    print("\n--------- MENÚ PRINCIPAL ---------")
    print("1. Registrar abogado")
    print("2. Mostrar abogados")
    print("3. Buscar abogado")
    print("4. Eliminar abogado")
    print("5. Estadísticas")
    print("6. Salir")

    opcion = input("Seleccione una opción: ")

    # ==========================================
    # REGISTRAR ABOGADO
    # ==========================================
    if opcion == "1":

        print("\n--- REGISTRO DE ABOGADO ---")

        # Diccionario vacío
        abogado = {}

        abogado["id"] = input("Ingrese el ID: ")
        abogado["nombre"] = input("Ingrese el nombre: ")
        abogado["especialidad"] = input("Ingrese la especialidad: ")
        abogado["experiencia"] = input("Años de experiencia: ")

        # IA sencilla
        años = int(abogado["experiencia"])

        if años >= 15:
            nivel = "Experto"
        elif años >= 7:
            nivel = "Profesional"
        else:
            nivel = "Principiante"

        abogado["nivel"] = nivel

        # Guardar en la lista
        losabogados.append(abogado)

        print("\n✅ Abogado registrado correctamente.")
        print("Nivel detectado por IA:", nivel)

    # ==========================================
    # MOSTRAR ABOGADOS
    # ==========================================
    elif opcion == "2":

        print("\n--- LISTA DE ABOGADOS ---")

        if len(losabogados) == 0:
            print("No hay abogados registrados.")
        else:

            contador = 1

            for abogado in losabogados:

                print("\nAbogado", contador)
                print("ID:", abogado["id"])
                print("Nombre:", abogado["nombre"])
                print("Especialidad:", abogado["especialidad"])
                print("Experiencia:", abogado["experiencia"])
                print("Nivel IA:", abogado["nivel"])

                contador += 1

    # ==========================================
    # BUSCAR ABOGADO
    # ==========================================
    elif opcion == "3":

        print("\n--- BUSCAR ABOGADO ---")

        buscar = input("Ingrese el nombre del abogado: ")

        encontrado = False

        for abogado in losabogados:

            if abogado["nombre"] == buscar:

                print("\n✅ Abogado encontrado")
                print("ID:", abogado["id"])
                print("Especialidad:", abogado["especialidad"])
                print("Experiencia:", abogado["experiencia"])
                print("Nivel IA:", abogado["nivel"])

                encontrado = True

        if encontrado == False:
            print("❌ No se encontró el abogado.")

    # ==========================================
    # ELIMINAR ABOGADO
    # ==========================================
    elif opcion == "4":

        print("\n--- ELIMINAR ABOGADO ---")

        eliminar = input("Ingrese el nombre del abogado: ")

        eliminado = False

        for abogado in losabogados:

            if abogado["nombre"] == eliminar:

                losabogados.remove(abogado)

                print("✅ Abogado eliminado.")
                eliminado = True

        if eliminado == False:
            print("❌ No existe ese abogado.")

    # ==========================================
    # ESTADÍSTICAS
    # ==========================================
    elif opcion == "5":

        print("\n--- ESTADÍSTICAS ---")

        total = len(losabogados)

        expertos = 0
        profesionales = 0
        principiantes = 0

        for abogado in losabogados:

            if abogado["nivel"] == "Experto":
                expertos += 1

            elif abogado["nivel"] == "Profesional":
                profesionales += 1

            else:
                principiantes += 1

        print("Total de abogados:", total)
        print("Expertos:", expertos)
        print("Profesionales:", profesionales)
        print("Principiantes:", principiantes)

    # ==========================================
    # SALIR
    # ==========================================
    elif opcion == "6":

        print("\nGracias por usar la IA de abogados.")
        break

    # ==========================================
    # OPCIÓN INVÁLIDA
    # ==========================================
    else:
        print("\n❌ Opción inválida.")
```
